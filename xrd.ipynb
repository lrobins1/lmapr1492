```python
from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator
from IPython.display import display

with MPRester("o1xT73gLJ9EAfjhPW2") as m:
    structure = m.get_structure_by_material_id("mp-14116")

    
    # création du calulcateur
calculateur=XRDCalculator()
pattern=calculateur.get_pattern(structure)
hkls=pattern.hkls
angles=pattern.x
intens=pattern.y
```

## Premier pic du diffractogramme

```python
print("Indices :", hkls[0])
print("Angle 2\u03B8 :", angles[0], "degrés")
print("Intensité :",intens[0], "%")
```

## Deuxième pic du diffractogramme

```python
print("Indices :", hkls[1])
print("Angle 2\u03B8 :", angles[1], "degrés")
print("Intensité :",intens[1], "%")
```

## Troisème pic du diffractogramme

```python
print("Indices :", hkls[3])
print("Angle 2\u03B8 :", angles[3], "degrés")
print("Intensité :",intens[3], "%")
```
